```python
# Import libraries
import pandas as pd
```

```python
# Create a list of words to pair with products
words = [
    'buy', 'price', 'clearance', 'discount', 'promotion', 'promo', 'shop', 'online-only', 
    'sale', 'bargain', 'savings', 'deal', 'offer', 'affordable', 'budget', 'value',
    'special', 'limited', 'exclusive', 'buy now', 'save', 'cost', 'slash', 'markdown',
    'coupon', 'cashback', 'incentive', 'buy one get one', 'free shipping', 'voucher',
    'wholesale', 'retail', 'purchase', 'shop now', 'best price', 'outlet', 'hot deal',
    'crazy deal'
]
print(words)
```

```python
products = [
    'sofas', 'convertible sofas', 'love seats', 'recliners', 'sofa beds', 'sectional sofas', 'armchairs', 
    'chaise lounges', 'futons', 'ottomans', 'bean bags', 'daybeds', 'loungers', 'benches', 
    'couches', 'accent chairs', 'club chairs', 'rocking chairs', 'swivel chairs', 'massage chairs', 
    'theater seating', 'gaming chairs', 'wingback chairs', 'bar stools', 'dining chairs', 
    'folding chairs', 'office chairs', 'desk chairs', 'ergonomic chairs', 'high-back chairs'
]
print(products)
```

```python
# Create an empty list keywords
keywords_list = []

# Loop through the products
for product in products:
    # loop through the words
    for word in words:
        # Append combinations
        keywords_list.append([product,product + ' ' + word])
        keywords_list.append([product, word + ' ' + word])

# View the keyword list
print(keywords_list)
```

```python
# Transform list to Dataframe
keywords_df = pd.DataFrame(keywords_list, columns = ['Ad Group', 'Keyword'])

# View the Dataframe
print(keywords_df)
```

```python
# Create a campaign column
keywords_df['Campaign'] = 'SEM_Sofas'

# Add a criterion column
keywords_df['Criterion Type'] = 'Exact'

# Save the final keywords to csv
keywords_df.to_csv('keywords.csv',index=False)
```
